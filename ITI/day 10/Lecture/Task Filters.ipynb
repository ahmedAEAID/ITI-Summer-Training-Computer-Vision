### 1-Filterâ€™s trackbar (gaussian,averaging,median,bilateral)

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def nothing(x):
    pass

img = cv2.imread('images/001.jpeg')
type(img)
```

```python

cv2.namedWindow('Filters')
cv2.createTrackbar('bar','Filters',0,100,nothing)
cv2.createTrackbar('mode','Filters',0,3,nothing)
font =cv2.FONT_HERSHEY_SIMPLEX
color_text =(100,100,100)
while True:
    img = cv2.imread('images/001.jpeg')
    ksize = cv2.getTrackbarPos('bar','Filters')
    mode  = cv2.getTrackbarPos('mode','Filters')
    if ksize %2 ==0:
        ksize += 1
    if mode ==0:
        cv2.putText(img, 'Averaging Filter',(10,50),font,1,color_text,2)
        img = cv2.blur(img,(ksize,ksize))
    elif mode == 1 :
        cv2.putText(img, 'Gaussian Filter',(10,50),font,1,color_text,2)
        img = cv2.GaussianBlur(img,(ksize ,ksize ),0)
    elif mode == 2:
        cv2.putText(img, 'Median Filtering',(10,50),font,1,color_text,2)
        img = cv2.medianBlur(img, ksize)
    else :
        cv2.putText(img, 'Bilateral Filter',(10,50),font,1,color_text,2)
        img = cv2.bilateralFilter(img, ksize, 75, 75)
    cv2.imshow('Filters',img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
    
```

### 2-search and implement (salt and piper noise)

```python

def add_salt_and_pepper_noise(image, prob): 
    
    output = image.copy()
    if len(image.shape) == 2: 
        black = 0
        white = 255            
    else:
        black = (0,0,0)
        white = (255,255,255)
    probs = np.random.random(output.shape[:2])
    output[probs < prob] = black
    output[probs > 1 - prob] = white
    return output
```

```python
img = cv2.imread('images/001.jpeg')

img = cv2.resize(img,None,fx=0.5,fy=0.5)
img2 = add_salt_and_pepper_noise(img,.01)
img3 = cv2.medianBlur(img2, 3)# -------to remove salt and pepper (median filter) more effiant--------

# img3 = cv2.blur(img2,(3,3)) 
# img3 = cv2.GaussianBlur(img2,(3,3 ),0) 

# in same window
# cv2.imshow('Filters',np.hstack((img,img2,img3)))

# in 3 different windows
cv2.imshow('orginal',img) # show orginal image 
cv2.imshow('image with noise',img2) # show image after adding salt and pepper noise 
cv2.imshow('after filter',img3)  # show image after Filter 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
