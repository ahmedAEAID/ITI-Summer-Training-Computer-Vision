```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

### Q1 load the following image and make red rectangle around face and write below rectangle “Cat” with yellow color.


```python
img_cat = cv2.imread('images/original.png')
img_cat1 = np.copy(img_cat)
cv2.rectangle(img_cat1, (800,200),(1500,800),(0,0,255),5)
font  = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img_cat1, 'Cat',(850,900),font,4,(0,255,255),6)
# cv2.imshow('Cat Face',img_cat1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
fig , ax = plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(img_cat[:,:,::-1]),ax[0].set_title('original')
ax[1].imshow(img_cat1[:,:,::-1]),ax[1].set_title('face cat')

plt.show()
```

### Q2 load the following image and apply sharpen filter.


```python
img_blured = cv2.imread('images/blurred.png')
img_blured = cv2.resize(img_blured,None, fx=.25,fy=.25)
kernel = np.array([
    [-1,-1,-1],
    [-1,9,-1],
    [-1,-1,-1]
])
img_sharp = cv2.filter2D(img_blured, -1,kernel,cv2.BORDER_DEFAULT)


## my observation
## ## after (resize) (blurred image) and used (sharpen filter) image is improved
## after (sharpen filter) - blur in image is hidden 


cv2.imshow('image blured before and after sharpen filter'
           ,np.hstack((img_blured,img_sharp)))

cv2.waitKey(0)
cv2.destroyAllWindows()

##------by matplotlib------
# fig , ax = plt.subplots(1,2,figsize=(10,10))
# ax[0].imshow(img_blured[:,:,::-1]),ax[0].set_title('blured')
# ax[1].imshow(img_sharp[:,:,::-1]),ax[1].set_title('After sharpen filter')
# plt.show()
```

### Q3 load the following image and apply blur filter 

```python

img_sharpen = cv2.imread('images/sharpen.png')
img_sharpen = cv2.resize(img_sharpen,None, fx=.5,fy=.5)
img_after_blur= cv2.blur(img_sharpen,(3,3),cv2.BORDER_DEFAULT)

## my observation
## after (resize) sharpen_image and used (blur filter) image is improved
## after (blur filter) - sharp in image is hidden 

cv2.imshow('image sharpen before and after blur filter',np.hstack((img_sharpen,img_after_blur)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
