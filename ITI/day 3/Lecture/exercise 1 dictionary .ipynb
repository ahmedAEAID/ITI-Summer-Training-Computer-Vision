## Write a Python program to drop empty Items from a given Dictionary.


```python
dict1 = {'c1': 'Red', 'c2': 'Green', 'c3':None}
print("Original Dictionary:")
print(dict1)
print("dropping empty items:")
dict1 = {k:v for k, v in dict1.items() if v is not None}
print(dict1)
```
