```python
#Q1 Create a Vehicle class with name, max_speed and mileage instance attributes.
class Vehicle :
    
#     Defines Attributes
    def __init__(self, name, max_speed, mileage):
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage

        
        
#   Q3  Add a function seating_capacity
    def seating_capacity(self,capacity):
        return  f'The seating capacity of a {self.name} is {capacity} passengers'

    
    
    
# Q2 Create a child class Bus that will inherit all of the 
# attributes and methods of the Vehicle class.

class Bus (Vehicle):
    
    
#     Q5 Add an instance variable to the Bus class called 
#    capacity and give it a default value of 50.

    def __init__ (self, name, max_speed, mileage):
      
        super().__init__( name, max_speed, mileage)
        self.capacity = 50
        
        
#     Q4 override the function seating_capacity in the Bus class 
#     to give the capacity argument a default value of 50.

    def seating_capacity(self,capacity=50):
        return super().seating_capacity(capacity)
    
    
#     Q6 Create a function fare()

    def fare(self):
        return self.capacity *100
```

```python
bus = Bus('HamdaBus',150,200)

print(bus.name)
print(bus.max_speed)
print(bus.mileage)
print(bus.capacity)
print(bus.fare())
print(bus.seating_capacity(200))
```
