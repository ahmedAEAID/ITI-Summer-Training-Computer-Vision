###  Q1- Create image 256*256 Draw red square 50*50 in the center of blue circle with radius 80 the center of both square and circle is the same center of image. 

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
```

```python
img = np.zeros((256,265,3),np.uint8)
center= np.array(img.shape[:-1])//2
r = 80
th = 3
cv2.circle(img,center,r,(0,0,255),th)
cv2.rectangle(img , center-25,center+25,(255,0,0),th)
plt.imshow(img)
plt.show()
```

### Q2 make trackbar to control line position ,color and thickness in image 500*500 rgb.

```python
img = np.zeros((500,500,3),np.uint8)


def nothing(x):
    pass

cv2.namedWindow('image')
cv2.createTrackbar('x_start','image',0,img.shape[0],nothing)
cv2.createTrackbar('y_start','image',0,img.shape[1],nothing)
cv2.createTrackbar('x_end','image',0,img.shape[0],nothing)
cv2.createTrackbar('y_end','image',0,img.shape[1],nothing)
cv2.createTrackbar('red','image',0,255,nothing)
cv2.createTrackbar('green','image',0,255,nothing)
cv2.createTrackbar('blue','image',0,255,nothing)
cv2.createTrackbar('th','image',1,20,nothing)
img = np.zeros((500,500,3),np.uint8)
while 1:
    img [:,:] =0
    x = cv2.getTrackbarPos('x_start','image')
    y = cv2.getTrackbarPos('y_start','image')
    x1 = cv2.getTrackbarPos('x_end','image')
    y1 = cv2.getTrackbarPos('y_end','image')
    r = cv2.getTrackbarPos('red','image')
    g = cv2.getTrackbarPos('green','image')
    b = cv2.getTrackbarPos('blue','image')
    th = cv2.getTrackbarPos('th','image')
    cv2.line(img,(x,y),(x1,y1), (r,g,b) , th )
    cv2.imshow('image',img[:,:,::-1])
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

### Q3 make trackbar to control position,radius and color of circle in image 500*500 rgb.

```python
img = np.zeros((500,500,3),np.uint8)


def nothing(x):
    pass

cv2.namedWindow('image')
cv2.createTrackbar('x','image',0,img.shape[0],nothing)
cv2.createTrackbar('y','image',0,img.shape[1],nothing)
cv2.createTrackbar('raduis','image',0,250,nothing)
cv2.createTrackbar('red','image',0,255,nothing)
cv2.createTrackbar('green','image',0,255,nothing)
cv2.createTrackbar('blue','image',0,255,nothing)
cv2.createTrackbar('th','image',0,20,nothing)
img = np.zeros((500,500,3),np.uint8)
while 1:
    img [:,:] =0
    x = cv2.getTrackbarPos('x','image')
    y = cv2.getTrackbarPos('y','image')
    raduis = cv2.getTrackbarPos('raduis','image')
    r = cv2.getTrackbarPos('red','image')
    g = cv2.getTrackbarPos('green','image')
    b = cv2.getTrackbarPos('blue','image')
    th = cv2.getTrackbarPos('th','image')
    cv2.circle(img,(x,y),raduis,(r,g,b),th)
    cv2.imshow('image',img[:,:,::-1])
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

### Q5 create image 800*800 rgb with red circle in center of image with radius 100 and make small blue circle with radius of 6 that move within circumference of the big circle every 1 second 10 degree. 

```python
img = np.zeros((800,800,3),np.uint8)
center= np.array(img.shape[:-1])//2
r = 100
th = 3

angle = 0
while True:
    
    img [:,:] =0
    x = int(center[0] + (r) * math.cos(math.radians(angle)))
    y = int(center[1] + (r) * math.sin(math.radians(angle)))
    cv2.circle(img,center,r,(255,0,0),th)
    cv2.circle(img ,(x,y),6 ,(0,0,255),-1)
    cv2.imshow('image',img[:,:,::-1])
    k = cv2.waitKey(1000)
    angle += 10
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```
