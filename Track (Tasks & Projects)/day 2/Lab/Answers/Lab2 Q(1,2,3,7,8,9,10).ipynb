```python
# Question 1

from datetime import datetime

year = datetime.now().year
birth_Year = int(input('Please enter your birth year: '))

print(f'You are {year - birth_Year } Years Old')
```

```python
# Question 2

num = int(input('Please Enter any Number: '))

print(f'The Sum of Natural Numbers from 1 to {num} = {sum(range(1,num+1))}')
print(f'The Sum of Natural Numbers from 1 to {num} = {sum(range(1,num+1))/num}')
```

```python
# Question 3

print("Enter 'q' for exit.\n\n")

while(True):
    num = input('Enter any number: ')
    if num == 'q' :
        print('\nDone')
        break;
    elif int(num) % 2 == 0:
        print(f'{num} is an even number.')
    elif int(num) % 2 == 1:
        print(f'{num} is an odd number.')
    else:
        print('Invalid Number, again !')
```

```python
# Question 7

print('Enter 0 for exit.\n\n')
result=''
while(True):
    c= int(input('Enter ASCII code: '))    
    result = result+ chr(c)
    print(result)
    if c==0:
        break;
```

```python
# Question 8

from math import pi

radius = float(input('Enter the Radius of a circle: '))
area = pi * (radius**2)
print(f'area of circle = {area}')
```
