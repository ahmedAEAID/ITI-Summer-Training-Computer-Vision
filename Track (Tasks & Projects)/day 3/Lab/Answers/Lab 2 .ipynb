### Question 4:

```python
lst = [2, 0, 3, 13, 76, 5, 8, 10, 1, 33, 42]

for num in lst :
    if num == 0:
        print(num,"Zero")
    elif (num %2 ==0) and (num < 10):
        print(num,"Even less than 10")
    elif (num %2 ==0) and (num > 10):
        print(num,"Even greater than 10")
    elif (num %2 !=0) and (num < 10):
        print(num ,"Odd less than 10")
    elif (num %2 !=0) and (num > 10):
        print(num,"Odd greater than 10")
    
```
