### Q4- Write a program to check a character is vowel or not.

### Q8 - Write a program that keep on accepting number from the user until user enters Zero. Display thesum and average of all the numbers.


```python
print('please enter zero to Exit. ')
sum = 0
count = 0
while True :
    num= int(input(f'Enter the number ({count+1}) : '))
    if num == 0:
        break
    sum= sum + num
    count= count +1
print('sum = ',sum)
average= sum/count
print('average = ',average)
```

### Q10- Make Method to Return the Maximum and Minimum item from the given list.

```python

def maximum_and_minimum(list1):
    max = list1[0]
    min = list1[0]
    for num in list1:
        if max < num:
            max= num
        else:
            min=num
    return max ,min
lst=[100,30,22,40,60,20,10,300]

print(maximum_and_minimum(lst))
```
