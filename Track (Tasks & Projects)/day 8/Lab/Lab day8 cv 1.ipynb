```python
# Make trackbar for rgb colors with window “colors”


img = np.zeros((512,512,3),np.uint8)

def nothing(x):
    pass
cv2.namedWindow('colors')
cv2.createTrackbar('r','colors',0,255,nothing)
cv2.createTrackbar('g','colors',0,255,nothing)
cv2.createTrackbar('b','colors',0,255,nothing)
while True:
    r = cv2.getTrackbarPos('r','colors')
    g = cv2.getTrackbarPos('g','colors')
    b = cv2.getTrackbarPos('b','colors')
    img[:]=[b,g,r]
    cv2.imshow('colors',img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
# - run video stream using your laptop camera and save frames when i press key “s” 
# close camera when i press “q” ,name of images should be “img.png +{counter}



cap = cv2.VideoCapture(0)
counter = 0
while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        counter +=1
        cv2.imshow('myCamera',frame)
        k = cv2.waitKey(40)
        if k == ord('q'):
            print('Done..')
            break
        elif k == ord('s'):
            cv2.imwrite(f'videos/img({counter}).png',frame)
            print(f'img{counter} is saved')
    else:
        break
cap.release()
cv2.destroyAllWindows()
        
```
