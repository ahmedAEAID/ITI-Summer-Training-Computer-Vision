```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = np.zeros((256,256),np.uint8)



def nothing(x):
    pass

cv2.namedWindow('image')
cv2.createTrackbar('col','image',0,255,nothing)
cv2.createTrackbar('row','image',0,255,nothing)
while 1:
    img [:,:] =0
    col = cv2.getTrackbarPos('col','image')
    row = cv2.getTrackbarPos('row','image')
    img [row,:] = 255
    img[:,col] = 255
    cv2.imshow('image',img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```
