### Q2- run video stream using your laptop camera and save frames when i press key “s” close camera when i press “q” ,name of images should be “img.png +{counter}”

```python
cap = cv2.VideoCapture(0)
counter = 0
while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        counter +=1
        cv2.imshow('myCamera',frame)
        k = cv2.waitKey(40)
        if k == ord('q'):
            print('Done..')
            break
        elif k == ord('s'):
            cv2.imwrite(f'videos/img({counter}).png',frame)
            print(f'img{counter} is saved')
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

### Q3- Make 100*100 rgb pixel image with black background and make diagonal of image with (x) color such that x is came from rgb trackbar.

```python
img = np.zeros((100,100,3),np.uint8)

def nothing(x):
    pass

cv2.namedWindow('diagonal color')
cv2.createTrackbar('red','diagonal color',0,255,nothing)
cv2.createTrackbar('green','diagonal color',0,255,nothing)
cv2.createTrackbar('blue','diagonal color',0,255,nothing)

while 1:
    r = cv2.getTrackbarPos('red','diagonal color')
    g = cv2.getTrackbarPos('green','diagonal color')
    b = cv2.getTrackbarPos('blue','diagonal color')
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if i==j:
                img[j,i]=(b,g,r)
    cv2.imshow('diagonal color',img)
    k = cv2.waitKey(1)
    if k== ord('q'):
        break
cv2.destroyAllWindows()
    
```

### Q4- create 100*100 rgb image with black background and make 1 vertical line and 1 horizontal line the color of both should be controlled spereat from 2 trackbar First one “horz” and second one is “vert” 

```python
img = np.zeros((100,100,3),np.uint8)

def nothing(x):
    pass

cv2.namedWindow('horz and vert color')
cv2.createTrackbar('red_horz','horz and vert color',100,255,nothing)
cv2.createTrackbar('green_horz','horz and vert color',0,255,nothing)
cv2.createTrackbar('blue_horz','horz and vert color',0,255,nothing)
cv2.createTrackbar('red_vert','horz and vert color',100,255,nothing)
cv2.createTrackbar('green_vert','horz and vert color',0,255,nothing)
cv2.createTrackbar('blue_vert','horz and vert color',0,255,nothing)

row = 50
col =50

while 1:
    r = cv2.getTrackbarPos('red_horz','horz and vert color')
    g = cv2.getTrackbarPos('green_horz','horz and vert color')
    b = cv2.getTrackbarPos('blue_horz','horz and vert color')
    r2 = cv2.getTrackbarPos('red_vert','horz and vert color')
    g2 = cv2.getTrackbarPos('green_vert','horz and vert color')
    b2 = cv2.getTrackbarPos('blue_vert','horz and vert color')
    img[row,:] = (b,g,r)
    img[:,col] = (b2,g2,r2)
    cv2.imshow('horz and vert color',img)
    k = cv2.waitKey(1)
    if k== ord('q'):
        break
cv2.destroyAllWindows()
```
